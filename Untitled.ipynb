```python
# ===============================================
# Mock Skills Test Solution: Student Performance
# ===============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------
# Part A: Base Computing
# -------------------------

def to_base(n, b):
    """Convert decimal integer n into base b (2 <= b <= 16)."""
    digits = "0123456789ABCDEF"
    if n == 0:
        return "0"
    result = ""
    while n > 0:
        result = digits[n % b] + result
        n //= b
    return result

# Load dataset
df = pd.read_csv("student_scores.csv")

# Compute average Math score (rounded to integer)
avg_math = int(df["Math"].mean())

print("=== Part A: Base Conversion ===")
print(f"Average Math Score: {avg_math}")
print(f"Binary: {to_base(avg_math, 2)}")
print(f"Octal: {to_base(avg_math, 8)}")
print(f"Hexadecimal: {to_base(avg_math, 16)}")
print()

# -------------------------
# Part B: NumPy Analysis
# -------------------------

scores = df[["Math", "Science", "English", "Electronics"]].to_numpy()

print("=== Part B: NumPy Statistics ===")
for i, col in enumerate(["Math", "Science", "English", "Electronics"]):
    col_data = df[col].dropna().to_numpy()  # drop NaN for Electronics
    print(f"{col} -> Mean: {np.mean(col_data):.2f}, "
          f"Median: {np.median(col_data):.2f}, "
          f"Std: {np.std(col_data):.2f}")

# Correlation matrix
corr_matrix = np.corrcoef(
    np.nan_to_num(scores, nan=np.nanmean(scores[:, 3])) , rowvar=False
)
print("\nCorrelation Matrix:\n", corr_matrix)
print()

# -------------------------
# Part C: Pandas Wrangling
# -------------------------

print("=== Part C: Pandas Wrangling ===")

# Fill missing Electronics scores (only STEM students)
mean_elec_stem = df.loc[df["Track"] == "STEM", "Electronics"].mean()
df["Electronics"] = df.apply(
    lambda row: mean_elec_stem if pd.isna(row["Electronics"]) and row["Track"] == "STEM" else row["Electronics"],
    axis=1
)

# Create Average column
df["Average"] = df[["Math", "Science", "English", "Electronics"]].mean(axis=1, skipna=True)

# Top 5 students
top5 = df.sort_values(by="Average", ascending=False).head(5)
print("Top 5 Students by Average:\n", top5[["Name", "Track", "Average"]])
print()

# -------------------------
# Part D: Visualization
# -------------------------

print("=== Part D: Visualizations ===")

# Average scores by track
track_avg = df.groupby("Track")[["Math", "Science", "English", "Electronics"]].mean()

plt.figure(figsize=(8, 5))
track_avg.plot(kind="bar", figsize=(10, 6))
plt.title("Average Scores by Track")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.legend(title="Subject")
plt.tight_layout()
plt.show()

# Boxplot of core subjects
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[["Math", "Science", "English"]])
plt.title("Distribution of Scores in Math, Science, and English")
plt.ylabel("Scores")
plt.tight_layout()
plt.show()

# Heatmap of correlation matrix
plt.figure(figsize=(6, 5))
sns.heatmap(df[["Math", "Science", "English", "Electronics"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Subjects")
plt.tight_layout()
plt.show()
```

```python
# ===============================================
# Mock Skills Test Solution 2: Sales & Customer Insights
# ===============================================

# Import required libraries
import pandas as pd              # For data wrangling and analysis
import numpy as np               # For numerical computations
import matplotlib.pyplot as plt  # For plotting
import seaborn as sns            # For advanced visualization

# -------------------------
# Part A: Base Computing
# -------------------------

def from_base(s, b):
    """
    Convert a string 's' in base 'b' (2 <= b <= 16) into a decimal integer.
    Example: from_base("1A3", 16) -> 419
    """
    digits = "0123456789ABCDEF"  # Allowed digits for bases up to 16
    s = s.upper()                # Make sure letters are uppercase (e.g., a -> A)
    value = 0                    # Initialize decimal value
    for char in s:               # Iterate through each character in the string
        value = value * b + digits.index(char)  # Multiply by base and add digit value
    return value

# Demonstration of base conversions
print("=== Part A: Base Conversion ===")
print(f"Binary 10110 -> {from_base('10110', 2)}")   # Convert binary to decimal
print(f"Octal 72 -> {from_base('72', 8)}")          # Convert octal to decimal
print(f"Hex 1A3 -> {from_base('1A3', 16)}")         # Convert hex to decimal
print()

# -------------------------
# Part B: NumPy
# -------------------------

# Load dataset into a Pandas DataFrame
df = pd.read_csv("sales_data.csv")

print("=== Part B: NumPy Statistics ===")

# Convert columns to NumPy arrays for numerical operations
quantities = df["Quantity"].to_numpy()
prices = df["Price"].to_numpy()
discounts = df["Discount"].fillna(0).to_numpy()  # Replace NaN with 0 for NumPy array

# Calculate mean and standard deviation for Quantity, Price, Discount
for arr, name in [(quantities, "Quantity"), (prices, "Price"), (discounts, "Discount")]:
    print(f"{name} -> Mean: {np.mean(arr):.2f}, Std: {np.std(arr):.2f}")

# Compute total sales per order using NumPy array arithmetic
# Formula: TotalSales = Quantity  Price  (1 - Discount)
total_sales_np = quantities * prices * (1 - discounts)

# Show the first 10 computed sales values
print("\nFirst 10 Total Sales (NumPy):", total_sales_np[:10])
print()

# -------------------------
# Part C: Pandas Wrangling
# -------------------------

print("=== Part C: Pandas Wrangling ===")

# Fill missing discounts (NaN) with the overall mean discount
df["Discount"] = df["Discount"].fillna(df["Discount"].mean())

# Create a new column 'TotalSales'
df["TotalSales"] = df["Quantity"] * df["Price"] * (1 - df["Discount"])

# Find top 3 customers by total sales (group by customer, sum sales, sort, take 3 highest)
top_customers = df.groupby("Customer")["TotalSales"].sum().sort_values(ascending=False).head(3)
print("Top 3 Customers by Total Sales:\n", top_customers, "\n")

# Convert OrderDate to datetime for time-based analysis
df["OrderDate"] = pd.to_datetime(df["OrderDate"])

# Compute monthly sales totals (group by month using Period "M")
monthly_sales = df.groupby(df["OrderDate"].dt.to_period("M"))["TotalSales"].sum()

print("Monthly Sales Trend:\n", monthly_sales, "\n")

# -------------------------
# Part D: Visualization
# -------------------------

print("=== Part D: Visualizations ===")

# ----- 1. Bar Chart: Average sales per region -----
region_avg = df.groupby("Region")["TotalSales"].mean()  # Group by Region and compute mean sales

plt.figure(figsize=(8, 5))                     # Create figure with size
region_avg.plot(kind="bar", color="skyblue", edgecolor="black")  # Plot bar chart
plt.title("Average Sales per Region")          # Chart title
plt.ylabel("Average Sales (USD)")              # Y-axis label
plt.xticks(rotation=0)                         # Keep x-axis labels horizontal
plt.tight_layout()                             # Adjust spacing
plt.show()

# ----- 2. Line Plot: Monthly Sales Trend -----
plt.figure(figsize=(10, 5))                    # Create figure
monthly_sales.plot(kind="line", marker="o", color="green")  # Line plot with markers
plt.title("Monthly Sales Trend (2024)")        # Title
plt.ylabel("Total Sales (USD)")                # Y-axis label
plt.xlabel("Month")                            # X-axis label
plt.grid(True, linestyle="--", alpha=0.7)      # Add gridlines
plt.tight_layout()                             # Adjust spacing
plt.show()

# ----- 3. Pie Chart: Sales by Product Type -----
product_sales = df.groupby("Product")["TotalSales"].sum()  # Group by product, sum sales

plt.figure(figsize=(6, 6))                     # Square figure for pie chart
plt.pie(product_sales, labels=product_sales.index, autopct="%1.1f%%", startangle=90)
plt.title("Sales Contribution by Product Type") # Title
plt.tight_layout()                             # Adjust spacing
plt.show()

```
